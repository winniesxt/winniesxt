Segmenting and Clustering Neighborhoods in Toronto Q1

1. build the code to scrape the following Wikipedia page, https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, in order to obtain the data that is in the table of postal codes and to transform the data into a pandas dataframe

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(url).text
Canada_data = BeautifulSoup(source, 'lxml')
table=Canada_data.find('table')
```

create the dataframe

```python
col_names = ['Postalcode','Borough','Neighborhood']
df= pd.DataFrame(columns = col_names)
```

```python
for tr_cell in table.find_all('tr'):
    row_data=[]
    for td_cell in tr_cell.find_all('td'):
        row_data.append(td_cell.text.strip())
    if len(row_data)==3:
        df.loc[len(df)] = row_data

df.head(12)
```

Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.


```python
indexNum = df[df['Borough'] == 'Not assigned'].index
df.drop(indexNum, inplace = True)

df.head(12)
```

```python
df1=df.groupby('Postalcode')['Neighborhood'].apply(lambda x: "%s" % ', '.join(x))
df1=df1.reset_index(drop=False)
df1.rename(columns={'Neighborhood':'Neighborhood_joined'},inplace=True)

df_merge = pd.merge(df, df1, on='Postalcode')

df_merge.drop(['Neighborhood'],axis=1,inplace=True)

df_merge.drop_duplicates(inplace=True)

df_merge.rename(columns={'Neighborhood_joined':'Neighborhood'},inplace=True)

df_merge.head(12)
```
